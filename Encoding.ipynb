```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
# Load dataset
wine = load_wine()

# Convert to DataFrame
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# Add target column
df['target'] = wine.target

df.head()
```

```python
df['target_cat'] = df['target'].map({0: 'class_A', 1: 'class_B', 2: 'class_C'})

df[['target', 'target_cat']].head()
```

```python
le = LabelEncoder()

df['label_encoded'] = le.fit_transform(df['target_cat'])

df[['target_cat', 'label_encoded']].head()
```

```python
ohe = OneHotEncoder(sparse_output=False)

encoded = ohe.fit_transform(df[['target_cat']])
encoded
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out())

encoded_df.head()
```

```python
df['quality'] = ['low', 'medium', 'high'] * (len(df)//3) + ['low']*(len(df)%3)

df[['quality']].head()
```

```python
oe = OrdinalEncoder(categories=[['low', 'medium', 'high']])

df['ordinal_encoded'] = oe.fit_transform(df[['quality']])

df[['quality', 'ordinal_encoded']].head()
```
